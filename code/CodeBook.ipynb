## In code/RDC.ipynb
- Installs and loads required packages (Python 3.10+)
- Imports raw datasets from data/raw
- Cleans and standardises variables across countries (Germany, Sweden, USA)
- Handles missing values and applies imputations where necessary
- Outputs cleaned datasets into data/clean as .csv files
- All results are reproducible by running cells sequentially

## Data Cleaning Decisions

The following cleaning procedures were applied across all datasets:

- Missing or empty observations were converted to `NA`
- Country names were standardized across datasets
- Observations were restricted to the years 2000–2023
- Only selected advanced economies were retained
- Data were converted into country-year panel format
- Numeric variables were converted to numeric data types where necessary
- Duplicate observations were removed where identified

COORD Imputation Assumption

The COORD variable is treated as a slow-moving institutional variable with limited variation over time.

For missing COORD observations:
- the nearest available previous year value was used (`forward fill`)
- if unavailable, the next available year value was used (`backward fill`)

This assumption is justified on the basis that labour market coordination institutions generally evolve gradually rather than changing abruptly between years.
